##### **Investigación de Overall**

$H_0:$ El "overall" promedio de jugadores brasilenos no es más alto que otros.

$H_1:$ El "overall" promedio de jugadores brasilenos es de hecho más alto que otros.

```python
df_p1.groupby('nationality')['overall'].mean().sort_values(ascending=False).to_frame().style.bar()
```

En términos del promedio, es obvio que, si ignoramos Puerto Rico como es un estado de los Estados Unidos, Brazil es el país latinamericano que tiene el promedio más alto. Sin embargo, necesitaremos chequear esta suposición.

Porque tenemos muestras muy pequeñitas para Costa Rica y las que están por debajo, excluimos esos países. Además, vemos que Puerto Rico con su "overall" promedio que es mejor que Brazil solo tiene un jugador solo de todos modos.

```python
latam_2 = [
    'Argentina',
    'Brazil',
    'Bolivia',
    'Chile',
    'Colombia',
    'Ecuador',
    'Mexico',
    'Paraguay',
    'Peru',
    'Uruguay',
    'Venezuela'
]
```

```python
df_p1 = df[df['nationality'].isin(latam_2)]
```

Ahora, examinamos a tráves de la prueba de Teacher los promedios de otros contra Brazil en "overall".

```python
from scipy.stats import ttest_ind
```

```python
data1 = df[df['nationality'] == 'Brazil']['overall']
h = []
for país in latam_2:
    if país == 'Brazil':
        continue
    else:
        data2 = df[df['nationality'] == país]['overall']
        stat, p = ttest_ind(data1, data2)
        h.append(p)
        print(país, ":", round(p, 4))
```

Ahora tenemos los p-valores de cada prueba entre Brazil y cada país individual con muestra suficiente. Si aglun p-valor es más que o igual a 0.05, no deberíamos rechazar el hípotesis nula. Los números se redondean, pero se redondean al cuarto decimal, lo que indica que estos números son como máximo valores-p de aproximadamente 0.0005.

Por formalidad, hacemos la siguiente prueba.

```python
if max(h) >= 0.05:
    print("No rechazamos lo nulo.")
else:
    print("Rechazamos lo nulo.")
```

Podemos decir, con más que 95% confianza, que según las estadísticas de nuestros datos, Brazil tiene un jugador promedio mejor. Tecnícamente, tenemos una probabilidad elevada de error del tipo II, pero los p-valores son tan pequeños que no necesitamos preocuparnos. Asímismo, la prueba de Teacher tiene la suposición de distribución normal, pero nuestros tomaños de datos son suficiente para evitar esa necesidad.

Podemos usar ANOVA para determinar si existe un efecto de las categorías.

```python
from scipy.stats import f_oneway
```

```python
def anova(x, y):
    datos = []
    for dato in list(df_p2[x].unique()):
        datos.append(list(df_p2[df_p2[x] == dato][y]))

    stat, p = f_oneway(*datos)
    print(x, ";", y, ":", 'stat=%.3f, p=%.3f' % (stat, p))
    if p > 0.05:
        print("Es probable que los promedios de los grupos determinadas por", x, "tengan los mismos promedios.\n")
    else:
        print("Es probable que los promedios de los grupos determinadas por", x, "tengan promedios diferentes.\n")
```

```python
predictores = ['nationality', 'club_name', 'league_name', 'player_positions', 'preferred_foot']
respuesta = 'value_eur'
for predictor in predictores:
    anova(predictor, respuesta)
```

Tenemos una prueba de Teacher aquí para nuestra lista de parejaas según "overall" que examina las medias entre grupos seperados por dos desviaciones estándares.

```python
def height_test(overall):
    std = df_n1[df_n1['overall'] == overall]['height_cm'].describe()['std']
    mean = df_n1[df_n1['overall'] == overall]['height_cm'].describe()['mean']
    
    data1 = df[(df_n1['overall'] == overall) & (df_n1['height_cm'] <= mean - std)]['value_eur']
    data2 = df[(df_n1['overall'] == overall) & (df_n1['height_cm'] >= mean + std)]['value_eur']
    
    suma = len(data1) + len(data2)

    print('Suma de entradas en muestra:', suma, '-', 'Proporción más bajo:', round(len(data1)/suma, 2))
    print('Promedio de más bajo:', int(data1.mean()), '-', 'Promedio de más alto:', int(data2.mean()))
    stat, p = ttest_ind(data1, data2)
    print('Valor-p:', round(p, 4), '\n')
```

```python
for overall in overall_común:
    height_test(overall)
```

Al contrario, parece que los hombres con menos estatura con la misma calidad gana más que sus hombres iguales. Pero este "hecho" no es una parte de nuestra hípotesis. Por eso, todo que decimos es que rechazamos lo nulo.
